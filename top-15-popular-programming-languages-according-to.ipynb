```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from collections import Counter
```

```python
df = pd.read_csv('/kaggle/input/stack-overflow-annual-developer-survey2022/so_data.csv')
```

```python
# replacing missing values with NaN float values
df.replace(['NaN', 'None'], np.nan, inplace=True)
df.dropna(inplace = True)
```

```python
lang_counter = Counter() # Creating counnter object to store individual languages
for l in df['LanguageHaveWorkedWith']:
    lang_counter.update(l.split(';')) # updaing counter with different langauges
```

```python
languages = []
popularity = []
# taking first 15 programming languages based on polularity
for l in lang_counter.most_common(15):
    languages.append(l[0])
    popularity.append(l[1])
```

```python
# plotting the horizental bar graph
plt.style.use('fivethirtyeight')
plt.title("Top 15 popular programming languages according to Stack Overflow survey data")
plt.xlabel("Number of people who are using this language")
plt.ylabel("Languages")
plt.barh(languages,popularity)
plt.tight_layout()
plt.show()
```
